```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def co2_footprint_per_day(average_fuel_consumption, emission_factor, average_stops, delivery_radius):
    # CO2 pro Lieferwagen pro Tag
    co2_per_van_per_day = (average_fuel_consumption * emission_factor) * average_stops
    
    # CO2 pro Lieferwagen pro Tag für den gesamten Flottenradius
    co2_per_van_per_day_total = co2_per_van_per_day * delivery_radius
    
    return co2_per_van_per_day_total

def total_co2_footprint(co2_per_van_per_day_total, num_vans):
    # CO2 pro Flotte pro Tag
    co2_total = co2_per_van_per_day_total * num_vans
    
    return co2_total

# Eingabewerte
average_fuel_consumption = 15.4  # Liter/km
emission_factor = 2.65  # kg CO2/Liter
average_stops = 40
delivery_radius = 50  # km
num_vans = 70

# Berechnungen
co2_per_van_per_day_total = co2_footprint_per_day(average_fuel_consumption, emission_factor, average_stops, delivery_radius)
total_co2 = total_co2_footprint(co2_per_van_per_day_total, num_vans)

# Ausgabe der Ergebnisse
print(f"CO2 pro Lieferwagen pro Tag für den gesamten Flottenradius: {co2_per_van_per_day_total:.2f} kg CO2/Tag")
print(f"CO2 pro Flotte pro Tag: {total_co2:.2f} kg CO2/Tag")
```

```python
# Funktion zur Berechnung des CO2-Fußabdrucks
def co2_footprint_per_day(average_fuel_consumption, emission_factor, average_stops, delivery_radius):
    # CO2 pro Lieferwagen pro Tag
    co2_per_van_per_day = (average_fuel_consumption * emission_factor) * average_stops
    
    # CO2 pro Lieferwagen pro Tag für den gesamten Flottenradius
    co2_per_van_per_day_total = co2_per_van_per_day * delivery_radius
    
    return co2_per_van_per_day_total

# Eingabewerte
average_fuel_consumption = 15.4  # Liter/km
emission_factor = 2.65  # kg CO2/Liter
average_stops = 40
delivery_radius = 50  # km

# Anzahl der Lieferwagen
num_vans = np.arange(1, 101, 1)

# Berechnung des CO2-Fußabdrucks für jede Anzahl von Lieferwagen
co2_footprints = [co2_footprint_per_day(average_fuel_consumption, emission_factor, average_stops, delivery_radius) * num_vans for num_vans in num_vans]

# Plot
plt.plot(num_vans, co2_footprints)
plt.title('CO2-Fußabdruck in Abhängigkeit von der Anzahl der Lieferwagen')
plt.xlabel('Anzahl der Lieferwagen')
plt.ylabel('CO2-Fußabdruck pro Tag (kg CO2/Tag)')
plt.grid(True)
plt.show()
```
